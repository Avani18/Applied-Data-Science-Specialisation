```python
#Artist layer demonstration

from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
from matplotlib.figure import Figure 
fig = Figure()
canvas = FigureCanvas(fig)

#Create 10000 random numbers
import numpy as np 
x = np.random.randn(10000)

#Create an axes artist 
ax = fig.add_subplot(111)

#Generate a histogram of the random numbers
ax.hist(x, 100)

#Add a title to the figure
ax.set_title("Normal distribution")

#Save the figure
fig.savefig('matplotlib_histogram.png')
```

```python
#Scripting layer demonstration

import matplotlib.pyplot as plt 
import numpy as np 

x = np.random.randn(10000)
plt.hist(x, 100)
plt.title("Normal distribution")
plt.savefig('matplotlib_histogram_scripting.png')
plt.show()
```

```python
import pandas as pd
#import xlrd

df_can = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)

print ('Data read into a pandas dataframe!')
```

```python
df_can.head()
```

```python
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(2)
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

```python
df_can.isnull().sum()
```

```python
df_can.head()
```

```python
df_can.set_index(['Country'], inplace = True)
df_can.head()
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
import matplotlib as mpl

df_can.loc("Haiti", years)
plt.title("Immigration from Haiti")
plt.ylabel("Number of immigrants")
plt.xlabel("Years")

plt.show()
```
